```python
#Pandas is a Python library used to store and work with data in a table-like format called a DataFrame.
import pandas as pd
#Tells pandas to read a CSV file named Salary_Data.csv.
#Stores the data from the CSV into a DataFrame called dataset.
dataset=pd.read_csv("Salary_Data.csv")
```

```python
dataset
```

```python
#I/p and O/p Splitting
#We are storing input: YearsExperience in independent variable
independent=dataset[["YearsExperience"]]
independent
```

```python
#We are storing Output: Salary in dependent variable
dependent=dataset[["Salary"]]
dependent
```

```python
#matplotlib is a Python library for creating visualizations like line graphs, bar charts, scatter plots, etc
#pyplot is a sub-module of matplotlib.It provides a collection of functions that offer a MATLAB-like interface for creating a 
#wide variety of static 2D plots and visualizations. 
import matplotlib.pyplot as plt
#Creates a scatter plot using the independent values on the x-axis and dependent values on the y-axis (YearsExperience vs Salary).
plt.scatter(independent,dependent)
#Sets the label for the x-axis as “YearsExperience” and makes the font size 10
plt.xlabel('YearsExperience',fontsize=10)
#Sets the label for the y-axis as “Salary” with font size 10.
plt.ylabel('Salary',fontsize=10)
#Displays the plot window with all the customizations done above.
plt.show()
```

```python
#sklearn, short for Scikit-learn, is one of the most popular Python libraries for machine learning
from sklearn.model_selection import train_test_split
#train_test_split splits your dataset into training and testing parts
#test_size=0.30 → 30% of the data goes into test, 70% into train
#random_state=0 - Fixes randomness so you get the same split every time (important for reproducibility)
Xtrain,Xtest,Ytrain,Ytest=train_test_split(independent,dependent,test_size=0.30,random_state=0)
Xtrain,Xtest,Ytrain,Ytest
```

```python
#Loads the LinearRegression model class
from sklearn.linear_model import LinearRegression
#Creates an empty Linear Regression model object called regressor
regressor=LinearRegression()
#**Trains the model** using the training data (`Xtrain` and `Ytrain`)
regressor.fit(Xtrain,Ytrain)
```

```python
weight=regressor.coef_
weight
```

```python
bias=regressor.intercept_
bias
```

```python
#Predicts the output (Salary) for test inputs (YearsExperience) using the trained model and stores the result in Y_pred.
Y_pred=regressor.predict(Xtest)
```

```python
#mports the **`r2_score` function** from `sklearn.metrics`, which is used to **measure the performance** of regression models.
from sklearn.metrics import r2_score
#Calculates the R² score by comparing the actual values (Ytest) and the predicted values (Y_pred)
r_score=r2_score(Ytest,Y_pred)
r_score
```

```python
#pickle is a Python module used to save (serialize) and load (deserialize) Python objects
import pickle
#Sets the name of the file((usually ends with .sav) where the trained model will be saved
filename="Finalized_Linear_model.sav"
#Saves the trained model to disk in binary format.'wb' means write binary.
#When you save a model using pickle, it turns the object (your model) into a special binary format that Python can understand later.
#dump means "save to a file"
####Open the file in binary write mode, and dump (save) the trained regressor model into it####
pickle.dump(regressor,open(filename,'wb'))
```

```python
#Opens the saved model file (Finalized_Linear_model.sav) in binary read mode ('rb')(read binary).
#Load Python object back from the file to Loaded_model variable
loaded_model=pickle.load(open(filename,'rb'))
result=loaded_model.predict([[15]])
result
```
